# Cuestionario 3
¿En qué se diferencian pivot_table() y groupby() (con una función de agregación)?

- La función groupby(), cuando se utiliza con funciones de agregación, suele producir resultados en el formato ‘largo’ (un Series por cada función de agregación), mientras que pivot_table() puede producir resultados en el formato ‘ancho’ (muchos Series por cada función de agregación).

# Cuestionario 4
¿Qué afirmaciones sobre concat() y merge() son ciertas?

- La función concat() es más adecuada para agrupar verticalmente los DataFrames. Así lo indica el hecho de que el parámetro **axis** sea 0 por defecto.

concat() puede fusionar DataFrames tanto a lo largo del eje de índices (axis=0) como del eje de columnas (axis=1). Sin embargo, teniendo en cuenta que el parámetro **axis** es 0 por defecto, parece más conveniente utilizar concat() para agrupar verticalmente varios DataFrames (uno encima de otro).

- La función merge() es más adecuada para ‘coser’ los DataFrames horizontalmente.

merge() puede fusionar DataFrames a lo largo del eje del índice o del eje de la columna (dependiendo de los nombres de índices y columnas que se pasen en los parámetros left_on y right_on), pero se utiliza sobre todo para fusionar dos DataFrames basados en valores de columnas (o índices) comunes.
