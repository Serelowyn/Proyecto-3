# Cuestionario 1
¿En qué consiste la ingeniería de características?

- Crear nuevas variables a partir de las existentes.

En el aprendizaje automático, el proceso de creación de nuevas columnas mediante el procesamiento de datos de las columnas originales del dataset se llama ingeniería de características. Las nuevas columnas tienen por objeto ayudar a revelar señales útiles del dataset a los algoritmos de aprendizaje automático, lo que mejora la calidad general del modelo.

# Cuestionario 2
¿Qué transformación no es probable que forme parte de la ingeniería de características?

- Convertir las fechas de los strings en el tipo de datos datetime.

Se trata de una transformación de tipo de datos y, por tanto, no es probable que ayude a mejorar la calidad del modelo. Extraer diferentes componentes temporales de estos valores como columnas separadas puede ser útil, y también puede considerarse como ingeniería de características.

# Cuestionario 4

- apply()

Esto nos permite aplicar una función personalizada a lo largo de un objeto Series o un eje del objeto DataFrame.

# Cuestionario 5
Quieres ejecutar una función personalizada llamada post_code_and_address_to_territory_group para calcular las categorías basadas en los valores de diferentes columnas. ¿Cómo llamarías a la función apply() para hacerlo?

- df.apply(post_code_and_address_to_territory_group, axis=1)

¡Correcto! Necesitamos establecer el parámetro axis a 1 si queremos que apply() pase los valores de fila de la función. Si hubiéramos puesto el parámetro axis a 0, apply() habría pasado los valores de la columna a la función.
