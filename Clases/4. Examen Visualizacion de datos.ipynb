# Cuestionario 1
¿Cuáles son las razones principales para crear gráficos?

- Analizar los datos.

Al visualizar los datos, podemos obtener información que sería imposible si solo leyéramos números y datos sin procesar.

- Comunicarnos con otras personas.

Las visualizaciones son una forma eficaz de comunicar los resultados del análisis. Es especialmente cierto cuando tu público no comparte el mismo nivel de conocimientos técnicos que tú, es decir, las partes interesadas y las personas encargadas de tomar decisiones.

- Presentar asociaciones complejas.

Las visualizaciones podrían ser muy útiles para revelar asociaciones ocultas entre las variables de los datos

# Cuestionario 2
¿Por qué hay que evitar malas visualizaciones?

- Pueden ser engañosas.

Las malas visualizaciones pueden plantear más preguntas que respuestas. Pueden provocar decisiones incorrectas debido a la falta de detalles o al cambio de enfoque. Esto sería muy inapropiado, ya que una comunicación buena y ética implica transparencia. Como profesional de los datos, tu deber es no cometer errores de visualización y ser capaz de reconocerlos para evitar que los demás te engañen.

# Cuestionario 3
Llamas a la función plot() para un objeto DataFrame. ¿Qué tipo de gráfico se mostrará?

- Gráfico de líneas.

plot() crea por defecto un gráfico de líneas para cada una de las columnas con datos numéricos.

# Cuestionario 5
Quieres analizar rápidamente la correlación entre todas las variables de un dataset. ¿Qué funciones utilizarías?

- scatter_matrix() y corr()

scatter_matrix() permite comprobar visualmente cualquier correlación compleja y corr() permite obtener estimaciones objetivas de las correlaciones

# Cuestionario 6
Quieres ver el crecimiento de la población en varios estados a lo largo de varios años. ¿Qué tipo de gráfico utilizarías?

- Gráfico de líneas.

Los gráficos de línea conectan secuencialmente cada punto de datos con una línea. Estas conexiones implican que existe una relación entre puntos de datos consecutivos. Poniendo los años en el eje X (en orden ascendente) y el valor correspondiente a la población en el eje Y, podemos crear un gráfico de línea para cada estado.

- Gráfico de barras.

La versión agrupada de los gráficos de barras nos viene bien. Poniendo los años en el eje X (en orden ascendente) y el valor correspondiente a la población en el eje Y, podemos crear un gráfico de barras agrupadas donde cada grupo de barras contiene la población de cada estado para un año como una barra separada en el grupo de barras.
